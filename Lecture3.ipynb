# LIST 
- A built-in data type that stores set of values enclosed in [].
- It can store elements of different types (integer, float, string, etc.)

```python
marks = [23,45,65,23,67]
print(marks)
print(type(marks))
print(marks[3])
```

```python
student = ["Bhavi",23,"Mumbai"]
print(student)
```

```python
# lists  are mutable(can be changed) whereas strings are immutable
student[1]=25
print(student)
```

```python
# length of list
print(len(student))
```

```python
student[5] # list can't be accessed beyond its range
```

## List Slicing
- similar to string slicing because list also have indexing
- can have negative indexes

  

```python
# list_name[starting-index:ending_index] ending index is not included
ls = [23,44,56,21,"hello"]
print(ls[3:5])
print(ls[:5]) # ls[0:5]
print(ls[2:]) # ls[2:len(ls)]
print(ls[-5:-1])
```

## List Methods
- list methods usually modify the object in place, while string methods always create and return a new string.
- This distinction comes from their design philosophy: lists are containers meant for dynamic editing, while strings are treated as fixed sequences of characters.
- Some list operations (like slicing nums[1:] or using + to concatenate) produce a new list, but methods like .append(), .remove(), .extend(), .sort(), .reverse() directly mutate the original object.

```python
ls = [2,1,3,4,5]
print(ls.append(6)) # adds one element at the end => [2,1,3,4,5,6], it does not return anything
print(ls)
```

```python
print(ls.sort()) # does not return anything
print(ls) # sorts in ascending order
ls.sort(reverse = True) # descending order
print(ls)
```

```python
ls.reverse() # reverses the list
print(ls) 
```

```python
ls = [3,4,5,6]
ls.insert(3,"hello") # insert element at index ls.insert(index,value)
print(ls)
```

```python
ls = [2,1,3,1]
ls.remove(1) # removes first occurence of element 
ls.pop(1) # removes element at index ls.pop(index)
print(ls)
```

# Tuples in Python
- A built-in data type that lets us create immutable sequences of values.

```python
tup = (1,2,3,4)
print(type(tup))
print(tup[0])
print(tup[1]) # you can access elements in a tuple, but can't change them
```

```python
tup[1] = 3 # not allowed
```

```python
tup1 = () # no element in tuple is also allowed
print(tup1)
tup2=(1) # here 1 is treated as int because elements in tuple are separated by commas
print(type(tup2))
tup3 = (1,)
print(type(tup3))
print(tup3)
```

## Tuple Methods

```python
tup = (2,1,3,1)
print(tup.index(1)) # returns index of first occurence tup.index(element)
print(tup.count(1)) # counts total occurences tup.count(element)
```
