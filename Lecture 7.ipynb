## Open, read & close File
`f = open("file_name", "mode")`<br>
`file_name = sample.txt, demo.docx`<br>
`mode = r: read,w: write` <br>
`data = f.read( )`<br>
`f.close( )` <br>

```python
f = open("sample.txt","r")
data = f.read()
print(data)
print(type(data))
f.close()
```

## Reading a file
- data = f.read( ) #reads entire file
- data = f.readline( ) #reads one line at a time

```python
f = open("sample.txt","r")
data = f.read(5) # 5 letters
print(data)
f.close()
```

```python
f = open("sample.txt","r")
data = f.readline() # 1st line "a blank line is because of \n at the end of line"
print(data)
data1 = f.readline()
print(data1)
f.close()
```

```python
f = open("sample.txt","r")
data = f.read() 
print(data)
data1 = f.readline() # after f.read() the cursor reaches the end and only see the \n that's why blank line
print(data1)
data2 = f.readline()
print(data2)
f.close()
f.close()
```

## Writing to a file
`f = open("demo.txt","w")` #overwrites the entire file <br>
`f.write( "this is a new line" )` #adds to the file <br>
`f = open( "demo.txt","a")` <br>
- If we open a file in 'a' or 'w' mode and that file doesn't exist, then Python will create a new file.

```python
f = open("demo.txt","w")
f.write("Over write by using 'w' mode")
f.close()
```

```python
f = open("demo.txt","a")
f.write("This line is added using 'a' mode") # you can use \n for new line
f.close()
```


- 'r+' - read+overwrite (ptr start) - notruncate <br>
- 'w+' - read+write (ptr start) - truncate <br>
- 'a+' - append (ptr end) - notruncate

## with Syntax
`with open( “demo.txt”,“a”) as f:` <br>
    
    data = f.read()
- with command automatically closes the file, don't need to write f.close()

```python
with open("demo.txt","r") as f:
    data = f.read()
    print(data)
```

```python
with open("demo.txt","w") as f:
    f.write("hello world")
```

```python
with open("practice.txt","w") as f:
    f.write('Hi everyone\nwe are learning file i/o\nusing Java\nI like programming in Java')
    # f.write("  ")
```

```python
with open("practice.txt","r") as f:
    data = f.read()
new_data = data.replace("Java","Python")
print(new_data)

with open("practice.txt","w")as f:
    f.write(new_data)
```

```python
with open("practice.txt","r") as f:
    data = f.read()
    if(data.find("learning")!=-1):
        print("found")
    else:
        print("not found")  
```

```python
def check_for_line():
    word = "learning"
    data = True
    line_no = 1
    with open("practice.txt","r") as f:
        while data:
            data = f.readline()
            if(word in data):
                print(line_no)
                return
            line_no+=1
    return -1
check_for_line()        
```

```python
with open("numbers.txt","r") as f:
    data = f.read()
    print(data)
    num = ""
    for i in range(len(data)):
        if(data[i]==","):
            print(int(num))
            num=""
        else:
            num+=data[i]
```

```python
count = 0
with open("numbers.txt","r") as f:
    data = f.read()
    nums=data.split(",")
    for val in nums:
        if(int(val) % 2==0):
            count+=1
print(count)
```
