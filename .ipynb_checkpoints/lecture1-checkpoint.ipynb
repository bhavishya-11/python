## Comments in Python

-  Official Comments: The official way to write comments in Python is using the hash symbol (#) for single-line comments.
-  De Facto Multi-line Comments: If a triple-quoted string is not assigned to a variable or used as a docstring, the Python interpreter treats it as a string literal that is not used, and therefore effectively ignores it during execution. This behavior leads to its common use as a multi-line comment, even though it's technically a string.

```python
# this is sinle line comment
"""
This block of text is often used as a multi-line comment.
It will be ignored by the interpreter because it's not assigned
to a variable or used as a docstring.
"""
print("Hello")
```

## Docstrings
In Python, triple quotes (either """ or ''') are primarily used to define multi-line strings and docstrings.
While they can effectively "appear" as multi-line comments because the Python interpreter ignores unassigned string literals, they are not officially designated as comments.

Multi-line Strings: Triple quotes allow you to define a string that spans multiple lines without needing explicit newline characters (\n).

```python
    multi_line_string = """This is a string
    that spans multiple
    lines."""
    print(multi_line_string)
```

Docstrings: When a triple-quoted string is placed immediately after a function, class, or module definition, it serves as a docstring, providing documentation for that element.

```python
    def my_function():
        """This is a docstring for my_function.
        It explains what the function does."""
        pass
    print(my_function.__doc__)
```

# Types of Operators
1. Arthimetic operator (+ , - , * , / , % , **,//)

```python
a = 4
b = 3
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b) # floor division Floor division automatically performs the "round to the nearest number in the negative infinity direction"
print(-a//b)
print(a**b)
print(a%b) # gives remainder
```

2. Relational / Comparison Operators ( == , != , > , < , >= , <= )

```python
a=4
b=5
print(a==b)
print(a>b)
print(a<b)
print(a!=b)
print(a>=b)
print(a<=b)
```

3. Assignment Operators ( = , +=, -= ,*= , /= , %= ,**= )

```python
a = 2
b = 3
c = 4
a+=1
b-=1
c*=4
print(a)
print(b)
print(c)
```

4. Logical Operators ( not , and , or )

```python
val1 = True
val2 = False
print(not(val1))
print(val1 and val2) # returns true when both condition are true
print(val1 or val2) # returns true when either one of the condition is true
```
